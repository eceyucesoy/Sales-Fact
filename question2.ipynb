```python
#f
feature = data[['TotalDrinkSales', 'TotalFoodSales', 'TotalNonConSales']]

from sklearn.preprocessing import StandardScaler
scaledFeature = StandardScaler().fit_transform(feature)

print(f"Mean: {scaledFeature.mean()}")
print(f"Std: {scaledFeature.std()}")

scaledDataFrame = pd.DataFrame(scaledFeature, columns = feature.columns)

print("\nSee that the correlation does not change when normalization is applied")
print("\nCorrelation of unscaled data")
print(feature.corr())
print("\nCorrelation of scaled data ")
print(scaledDataFrame.corr())

print("\nCompute PCA")

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
z = pca.fit_transform(scaledFeature)

print(z)
print("\nData shape after PCA")
print(z.shape)

variance = pca.explained_variance_ratio_
print(f"Proportion of Variance: {variance}")
print("\nPCA is able to explain 90% of the variance in the original dataset")

print("\nPCA components")
print(pca.components_)

print(f'\nUnit: {pow(pca.components_[0,0],2) + pow(pca.components_[0,1],2)}')

# The total variance is Var(X1) + Var(X2) = n, which is 2
print(f'Eigenvalues: {pca.explained_variance_}')

z_dataframe = pd.DataFrame(z, columns=['PC1', 'PC2'])

corr_data = pd.concat([scaledDataFrame, z_dataframe], axis=1)

print("\nwe can see that the principal components are not correlated")
corr_data.corr()
```

```python
#g
feature = data[['TotalDrinkUnitSales', 'TotalFoodUnitSales', 'TotalNonConUnitSales']]

featureDataFrame = pd.DataFrame(feature, columns = feature.columns)

print("\nCompute PCA")

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
z = pca.fit_transform(feature)

print(z)
print("\nData shape after PCA")
print(z.shape)

variance = pca.explained_variance_ratio_
print(f"Proportion of Variance: {variance}")
print("\nPCA is able to explain 98,5% of the variance in the original dataset")

print("\nPCA components")
print(pca.components_)

print(f'\nUnit: {pow(pca.components_[0,0],2) + pow(pca.components_[0,1],2)}')

# The total variance is Var(X1) + Var(X2)  = n, which is 2
print(f'Eigenvalues: {pca.explained_variance_}')

z_dataframe = pd.DataFrame(z, columns=['PC1', 'PC2'])

corr_data = pd.concat([featureDataFrame, z_dataframe], axis=1)

print("\nwe can see that the principal components are not correlated")
corr_data.corr()
```
